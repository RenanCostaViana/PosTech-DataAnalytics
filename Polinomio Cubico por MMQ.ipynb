# Proposta

Você foi recém chamado(a) para um processo seletivo pela empresa de consultoria FIAPMETRICS.

A vaga consiste em integrar a equipe de Advanced Analytics, e foi proposto um teste técnico de nivelamento para que a empresa veja como podem avançar com você.

O teste consiste em encontrar o polinômio cúbico através do método dos mínimos quadrados (MMQ) que passa pelos pontos (-1, -10), (0, 5), (1, 6), (2, 11) e (3, 42).

Utilize como base a equação geral do polinômio cúbico:

$$ AX^3 + BX^2 + CX + D $$

Plot o gráfico demonstrando esse polinômio cúbico que melhor se ajusta aos pontos dados.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1, y1 = -1, -10
x2, y2 = 0, 5
x3, y3 = 1, 6
x4, y4 = 2, 11
x5, y5 = 3, 42

x = [x1, x2, x3, x4, x5]
y = [y1, y2, y3, y4, y5]

plt.figure(figsize = (10, 5))

plt.scatter(x1, y1, color = 'green', label = 'Pares ordenados (x,y)')
plt.scatter(x2, y2, color = 'green')
plt.scatter(x3, y3, color = 'green')
plt.scatter(x4, y4, color = 'green')
plt.scatter(x5, y5, color = 'green')

plt.plot(x, y, label = 'Gráfico que liga os pontos')
plt.grid(True)
plt.box(False)
plt.legend()
plt.title('Interpolação da reta sobre todos os pontos')

plt.show()
```

```python
A = np.array([[-1, 1, -1, 1], [0, 0, 0, 1], [1, 1, 1, 1], [8, 4, 2, 1], [27, 9, 3, 1]])
Y = np.array([[y1], [y2], [y3], [y4], [y5]])
```

```python
(a, b, c, d) = np.dot(np.dot(np.linalg.pinv(np.dot(A.T, A)), A.T), Y)
print(a, b, c, d)
```

```python
plt.figure(figsize = (10, 5))

plt.scatter(x1, y1, color = 'green', label = 'Pares ordenados (x,y)')
plt.scatter(x2, y2, color = 'green')
plt.scatter(x3, y3, color = 'green')
plt.scatter(x4, y4, color = 'green')
plt.scatter(x5, y5, color = 'green')

v_x = np.linspace(-1, 3)
plt.plot(x, y, label = 'Gráfico que liga os pontos')
plt.plot(v_x,
         (a * (v_x**3)) + (b * (v_x**2)) + (c * (v_x**1)) + d,
         label = 'Reta otimizada pelo MMQ')
plt.grid(True)
plt.box(False)
plt.legend()
plt.title('Interpolação da reta sobre todos os pontos')

plt.show()
```
