## 3 Limpe o dataset, removendo linhas duplicadas e tratando valores faltantes

```python
dados_dropados = dados.drop_duplicates()
```

```python
dados_dropados = dados_dropados.drop(['fixed acidity','citric acid','residual sugar','chlorides','free sulfur dioxide',
                             'total sulfur dioxide','density','pH','sulphates'], axis=1)
```

## 1 Gráficos refeitos com Seaborn

```python
import seaborn as sns
```

### 1.1 Distribuição de acidez volátil por qualidade do vinho

```python
ax = sns.lmplot(data = dados_dropados, x = 'quality', y = 'volatile acidity')

ax.fig.set_size_inches(10,6)
ax.ax.grid(axis = 'y', linestyle = '--')

plt.xlabel('Nível de Qualidade')
plt.ylabel('Acidez Volátil')
plt.title('Acidez Volátil x Nível de Qualidade')

plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=dados_dropados, x='quality', y='volatile acidity')

ax.grid(axis = 'y', linestyle = '--')

plt.xlabel('Nível de Qualidade')
plt.ylabel('Acidez Volátil')
plt.title('Acidez Volátil x Nível de Qualidade')

plt.show()
```

### 4.2 Distribuição do teor alcóolico por qualidade do vinho

```python
ax = sns.lmplot(data=dados_dropados, x='quality', y='alcohol')

ax.fig.set_size_inches(10,6)
ax.ax.grid(axis = 'y', linestyle = '--')

plt.xlabel('Nível de Qualidade')
plt.ylabel('Teor Alcoólico')
plt.title('Teor Alcoólico x Nível de Qualidade')

plt.show()
```

### 4.3 Relação entre acidez volátil e teor alcóolico

```python
ax = sns.lmplot(data=dados_dropados, x='alcohol', y='volatile acidity')

ax.fig.set_size_inches(10,6)

plt.xlabel('Teor Acoólico')
plt.ylabel('Acidez Volátil')
plt.title('Acidez Volátil x Teor Acoólico')
plt.grid(linestyle = '--')

plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=dados_dropados.groupby('alcohol').mean(), x='alcohol', y='volatile acidity')

ax.grid(linestyle = '--')

plt.xlabel('Teor Acoólico')
plt.ylabel('Acidez Volátil')
plt.title('Acidez Volátil x Teor Acoólico')

plt.show()
```

### 4.4 Relação entre teor alcóolico e qualidade do vinho

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=dados_dropados, x='quality', y='alcohol')

ax.grid(axis = 'y', linestyle = '--')

plt.xlabel('Nível de Qualidade')
plt.ylabel('Teor Alcoólico')
plt.title('Teor Alcoólico x Nível de Qualidade')

plt.show()
```

## 2 Dados Complementares

Comparação do teor alcoólico entre os vinhos tintos da Vinho Verde e os 5 melhores do mundo de acordo com o Taste Atlas [1]. Seus teores alcoólicos foram obtidos em suas páginas no site Wine Searcher [2].


[1] https://www.tasteatlas.com/best-rated-red-wines-in-the-world#bestProducts

[2] https://www.wine-searcher.com/

```python
top_5 = {"Château d'Armailhac 2006": 13.0,
          "Quails' Gate The Connemara": 14.0,
          "Inniskillin Cabernet Sauvignon Icewine": 9.5,
          "Jackson-Triggs Niagara Estate Reserve Meritage 2016": 12.5,
          "Grans Muralles": 14.5}
```

```python
top_5 = pd.Series(top_5)
top_5 = pd.DataFrame(top_5)
top_5.reset_index(inplace=True)
top_5.columns = ['Vinho', 'Teor Alcoólico']
```

```python
ax = sns.lmplot(data = dados_dropados, x ='quality', y ='alcohol')

ax.fig.set_size_inches(10,6)

plt.xlabel('Nível de Qualidade')
plt.ylabel('Teor Alcoólico')
plt.title('Teor Alcoólico x Nível de Qualidade')
plt.grid(axis = 'y', linestyle = '--')
plt.hlines(top_5['Teor Alcoólico'].mean(), xmin = 0, xmax = 8, colors ='Red')
plt.annotate('Média dos 5 melhores vinhos', (-0.3, float(top_5['Teor Alcoólico'].mean() + 0.1)), color = 'Red')

plt.show()
```
