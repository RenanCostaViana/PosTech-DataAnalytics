## Por Estado

```python
idhm_estados = dados[['Nome da Unidade da Federação','IDHM','IDHM Educação','IDHM Longevidade','IDHM Renda']]
```

```python
idhm_estados.head()
```

```python
''' def grafico_estado (coluna):
  idhm_estados.groupby('Nome da Unidade da Federação')[coluna].mean().sort_values(ascending=False).plot(kind='bar')
  plt.title(f'{coluna} por Estado')
  plt.xlabel('Estado')
  plt.ylabel(f'{coluna}')
  plt.show() '''
```

```python
idhm_estados_agrupado = idhm_estados.groupby('Nome da Unidade da Federação').mean().sort_values(by='IDHM',
    ascending=False)
idhm_estados_agrupado.head()
```

```python
plt.figure(figsize=(11, 6))

plt.scatter(idhm_estados_agrupado.index, idhm_estados_agrupado['IDHM'], label='IDHM')
plt.scatter(idhm_estados_agrupado.index, idhm_estados_agrupado['IDHM Educação'], label='IDHM Educação')
plt.scatter(idhm_estados_agrupado.index, idhm_estados_agrupado['IDHM Longevidade'], label='IDHM Longevidade')
plt.scatter(idhm_estados_agrupado.index, idhm_estados_agrupado['IDHM Renda'], label='IDHM Renda')

plt.title('IDHM e seus subtipos por Estado')
plt.legend(loc='lower left')
plt.xticks(rotation = 90)
plt.show()
```

## Por região

```python
região = {'Acre':'Norte', 'Amapá':'Norte', 'Amazonas':'Norte', 'Pará':'Norte', 'Rondônia':'Norte',
          'Roraima':'Norte', 'Tocantins':'Norte', 'Maranhão':'Nordeste', 'Piauí':'Nordeste',
          'Ceará':'Nordeste', 'Rio Grande do Norte':'Nordeste', 'Paraíba':'Nordeste',
          'Pernambuco':'Nordeste', 'Alagoas':'Nordeste', 'Sergipe':'Nordeste', 'Bahia':'Nordeste',
          'Minas Gerais':'Sudeste', 'Espírito Santo':'Sudeste', 'Rio de Janeiro':'Sudeste',
          'São Paulo':'Sudeste', 'Paraná':'Sul', 'Santa Catarina':'Sul', 'Rio Grande do Sul':'Sul',
          'Mato Grosso do Sul':'Centro-Oeste', 'Mato Grosso':'Centro-Oeste',
          'Goiás':'Centro-Oeste', 'Distrito Federal':'Centro-Oeste'}
```

```python
idhm_regiao = idhm_estados.copy()
```

```python
idhm_regiao['Região'] = idhm_regiao['Nome da Unidade da Federação'].map(região)
```

```python
idhm_regiao.head()
```

```python
''' def grafico_regiao (coluna):
  idhm_regiao.groupby('Região')[coluna].mean().sort_values(ascending=False).plot(kind='bar')
  plt.title(f'{coluna} por Região')
  plt.xlabel('Região')
  plt.ylabel(f'{coluna}')
  plt.show() '''
```

```python
idhm_regiao_agrupado = idhm_regiao.groupby('Região').mean(numeric_only=True).sort_values(by='IDHM', ascending=False)
idhm_regiao_agrupado
```

```python
plt.figure(figsize=(11, 6))

plt.scatter(idhm_regiao_agrupado.index, idhm_regiao_agrupado['IDHM'], label='IDHM')
plt.scatter(idhm_regiao_agrupado.index, idhm_regiao_agrupado['IDHM Educação'], label='IDHM Educação')
plt.scatter(idhm_regiao_agrupado.index, idhm_regiao_agrupado['IDHM Longevidade'], label='IDHM Longevidade')
plt.scatter(idhm_regiao_agrupado.index, idhm_regiao_agrupado['IDHM Renda'], label='IDHM Renda')

plt.title('IDHM e seus subtipos por Região')
plt.legend(loc='lower left')
plt.xticks(rotation = 90)
plt.show()
```

## Estados por região

```python
''' def grafico_estado_regiao (coluna, regiao):
  idhm_regiao.query('Região == @regiao').groupby('Nome da Unidade da Federação')[coluna].mean().sort_values(ascending=False).plot(kind='bar')
  plt.title(f'{coluna} por estado ({regiao})')
  plt.xlabel('Estado')
  plt.ylabel(f'{coluna} ')
  plt.show() '''
```

```python
def scatter_estado_regiao (regiao):
  regiao_agrupada = idhm_regiao.query('Região == @regiao').groupby('Nome da Unidade da Federação').mean(numeric_only=True).sort_values(by='IDHM', ascending=False)

  plt.figure(figsize=(11, 6))

  plt.scatter(regiao_agrupada.index, regiao_agrupada['IDHM'], label='IDHM')
  plt.scatter(regiao_agrupada.index, regiao_agrupada['IDHM Educação'], label='IDHM Educação')
  plt.scatter(regiao_agrupada.index, regiao_agrupada['IDHM Longevidade'], label='IDHM Longevidade')
  plt.scatter(regiao_agrupada.index, regiao_agrupada['IDHM Renda'], label='IDHM Renda')

  plt.title(f'IDHM e seus subtipos por estado ({regiao})')
  plt.legend(loc='lower left')
  plt.xticks(rotation = 90)
  plt.show()
```

```python
scatter_estado_regiao('Norte')
```

```python
scatter_estado_regiao('Nordeste')
```

```python
scatter_estado_regiao('Sudeste')
```

```python
scatter_estado_regiao('Sul')
```

```python
scatter_estado_regiao('Centro-Oeste')
```

## Abertura dos Indicadores

### Educação

```python
idhm_educacao = dados[['Nome da Unidade da Federação','IDHM Educação',
                       '% de 18 anos ou mais com fundamental completo', '% de 5 a 6 anos na escola',
                       '% de 11 a 13 anos nos anos finais do fundamental ou com fundamental completo',
                       '% de 15 a 17 anos com fundamental completo', '% de 18 a 20 anos com médio completo']]
```

```python
idhm_educacao.head()
```

```python
idhm_educacao_agrupado = idhm_educacao.groupby('Nome da Unidade da Federação').mean().sort_values(by='IDHM Educação',
    ascending=False)
idhm_educacao_agrupado.head()
```

```python
plt.figure(figsize=(11, 6))

plt.scatter(idhm_educacao_agrupado.index, idhm_educacao_agrupado['IDHM Educação']*100, label='IDHM Educação * 100')
plt.scatter(idhm_educacao_agrupado.index, idhm_educacao_agrupado['% de 18 anos ou mais com fundamental completo'],
            label='% 18+ com fundamental')
plt.scatter(idhm_educacao_agrupado.index, idhm_educacao_agrupado['% de 5 a 6 anos na escola'], label='% 5-6 na escola')
plt.scatter(idhm_educacao_agrupado.index,
              idhm_educacao_agrupado['% de 11 a 13 anos nos anos finais do fundamental ou com fundamental completo'],
              label='% 11-13 final fundamental')
plt.scatter(idhm_educacao_agrupado.index,
              idhm_educacao_agrupado['% de 15 a 17 anos com fundamental completo'], label='% 15-17 com fundamental')
plt.scatter(idhm_educacao_agrupado.index,
              idhm_educacao_agrupado['% de 18 a 20 anos com médio completo'], label='% 18-20 com médio')

plt.title('IDHM Educação e os índices que o compõe')
plt.legend(loc='lower left')
plt.xticks(rotation = 90)
plt.show()
```

### Longevidade

```python
idhm_longevidade = dados[['Nome da Unidade da Federação','IDHM Longevidade','Esperança de vida ao nascer']]
```

```python
idhm_longevidade.head()
```

```python
idhm_longevidade_agrupado = idhm_longevidade.groupby('Nome da Unidade da Federação').mean().sort_values(by='IDHM Longevidade',
    ascending=False)
idhm_longevidade_agrupado.head()
```

```python
idhm_longevidade_agrupado.plot(kind='bar', y='Esperança de vida ao nascer', figsize=(11, 6))
plt.title('Esperança de vida ao nascer por Estado')
plt.ylabel('Esperança de vida ao nascer')
plt.show()
```
